```python
import pandas as pd
import os
```

## OTD Data Process

```python
Path = r'E:\_Projects\Delivery Perforamce\\'
```

```python


Product = pd.read_csv(Path+'\\Input\\class_offerpf_map.csv')
Source_1 = pd.read_csv(Path+'\\Input\\Source 1.csv')
Source_2 = pd.read_csv(Path+'\\Input\\Source 2.csv')


Source = pd.concat([Source_1,Source_2],axis=0)

Source['Class'] = Source['Class'].astype('str')
Product['Class'] = Product['Class'].astype('str')

Data = pd.merge(Source, Product, on='Class',how='left')


ATM = Data.loc[(Data['Prod Grp']=='ATM') & (Data['Inventory Transaction Type Identifier']==62)]

ATM['Product Group'] = ATM['Prod Grp']

ATM = ATM.drop_duplicates(subset=['Inventory Item Serial Number'],keep='first')

ATM['Site'] = 'CHE'

ATM = ATM[['PO','Product Group','Class','MCID','PID Type','Actual Ship Date','Ship Period','Qty','IO Booked Dt(First)','IO Booked Dt(Last)','Booked Dt','Promise Dt (first)','Promise Dt (last)','Receipt Dt','PO Creation Dt','Delivery vs Orig PD','Compliance','Aging','Act Lt','Act Lt Wks','Act LT Wks Oride','System LT','System LT Wks','Act LT vs System LT','Factory LT','Factory LT Wks','Inventory Item Serial Number','Tracer Number','Manufactured Date','Offer PF']]
```

#### Update from EUMOCP

```python
Consolidation_Path = Path+'Consolidation\\'


dfs = []

for file_name in os.listdir(Consolidation_Path):
    if file_name.endswith('.csv'):
        
        print(file_name)
        
        df = pd.read_csv(os.path.join(Consolidation_Path, file_name))
        dfs.append(df)

Consolidation = pd.concat(dfs, ignore_index=True)

Consolidation.to_csv(Consolidation_Path+'Consolidate.csv',index=False)
```

```python
Consolidation = Consolidation[['Org Code', 'Region', 'Customer_End Customer','Booked Dt','PID','IO No.','IO SSD', 'IO BOOKED DATE','filref']]
Consolidation.drop_duplicates(subset=['PID','IO No.', 'IO BOOKED DATE', 'IO SSD'], inplace=True)
Consolidation = Consolidation[Consolidation['IO No.'].notnull()]
Consolidation = Consolidation[Consolidation['IO SSD'].notnull()]
val = {'IO No.':'PO','PID':'MCID'}
Consolidation.rename(columns= val, inplace=True)
```

```python
first_dt = Consolidation.copy()
last_dt = Consolidation.copy()

first_dt.sort_values(by=['filref'])
first_dt = Consolidation[Consolidation['IO SSD'].notnull()]
first_dt.drop_duplicates(subset=['MCID','PO'], keep='first', inplace=True)

last_dt.sort_values(by=['filref'])
last_dt.drop_duplicates(subset=['MCID','PO'], keep='last', inplace=True)
```

```python
Data = pd.merge(ATM , first_dt, on=['MCID','PO'], how='left')
Data = pd.merge(Data , last_dt, on=['MCID','PO'], how='left')
```

```python
Data['IO Booked Dt(First)'] = Data['IO BOOKED DATE_x']
Data['IO Booked Dt(Last)'] = Data['IO BOOKED DATE_y']
Data['IO Booked Dt(Last)'] = Data['IO BOOKED DATE_y']
Data['Booked Dt'] = Data['Booked Dt_y']
Data['Promise Dt (first)'] = Data['IO SSD_x']
Data['Promise Dt (last)'] = Data['IO SSD_y']
Data.rename(columns={'Region_x':'Region'}, inplace=True)
Data['Customer Name'] = Data['Customer_End Customer_x']
Data['Country'] = Data['Org Code_x']
```

```python
Data.rename(columns={'Actual Ship Date':'SHIPPED Dt'}, inplace=True)

#Data.to_csv(r'C:\Vasanthan\Report\Delivery Performance\Automation\Data.csv',index=False)
```

```python
# Convert each column to datetime format
date_columns = [
    'SHIPPED Dt',
    'IO Booked Dt(First)',
    'IO Booked Dt(Last)',
    'Booked Dt',
    'Promise Dt (first)',
    'Promise Dt (last)',
    'Manufactured Date'
]

for col in date_columns:
    Data[col] = pd.to_datetime(Data[col])
```

```python
#Data.to_csv(r'C:\Vasanthan\Report\Delivery Performance\Automation\Data.csv',index=False)
```

### Caculation

```python
Data['Delivery vs Orig PD'] = (Data['SHIPPED Dt'] - Data['Promise Dt (first)']).dt.days

Data.loc[(Data['Delivery vs Orig PD'] < 1) | (Data['Delivery vs Orig PD'].isnull()), 'Compliance'] = 'On-Time'

Data['Compliance'] = Data['Compliance'].fillna('Late')
```

```python
def categorize_delivery(days):
    if days < 0:
        return "Early"
    elif days == 0:
        return "On-Time"
    elif days <= 5:
        return "1) 1-5 Days Late"
    elif days <= 10:
        return "2) 5-10 Days Late"
    elif days <= 20:
        return "3) 10-20 Days Late"
    elif days <= 30:
        return "4) 20-30 Days Late"
    else:
        return "5) 30+ Days Late"

Data['Delivery Category'] = Data['Delivery vs Orig PD'].apply(categorize_delivery)
Data['Site'] = 'CHE'
```
