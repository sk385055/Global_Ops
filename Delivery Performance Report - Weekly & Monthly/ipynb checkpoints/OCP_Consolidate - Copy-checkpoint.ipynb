```python
import os
import pandas as pd
import glob
from pyxlsb import open_workbook as open_xlsb
```

```python
folder_path = r'E:\_Projects\_outputs\orders\snd' # Replace with the actual folder path
```

```python
# Get a list of all files in the folder
files = glob.glob(folder_path+'\workbench_data_eumocp_wPO_2024????_v2.csv')

files.sort()
```

```python
consolidated_data = pd.DataFrame()

for file in files:
    print(file)
    df=pd.read_csv(file, encoding='utf-8', low_memory=False, keep_default_na='')
    #df= df[df['Plant'].isin(['CHE','CHS','EN2','JB2'])]
    #df= df[df['Plant'].isin(['EN2','JB2'])]
    df= df[df['PID Type'].isin(['M'])]
    

    df['filref'] = os.path.basename(file)
    df = df[['PO No.','Org Code','Plant','Region','Customer_End Customer','DEMAND VAL','Booked Dt','Ship Date','Product Range','Class','PID','PID Desc','PID Type','IO No.','IO BOOKED DATE','IO SSD','filref']]
    consolidated_data = pd.concat([consolidated_data,df],axis=0)
    
```

```python
consolidated_data.drop_duplicates(inplace=True)
```
