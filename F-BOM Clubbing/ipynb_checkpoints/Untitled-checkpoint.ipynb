```python
import pandas as pd
import xlrd
from openpyxl import load_workbook

# Load the Excel file and the specific sheet named 'PPL Sets'
file_path = "E:\\_Projects\\FBOM\\Input\\2062.xlsx"
sheet_name = 'Combinations and Defaults'

# Load the workbook using xlrd
workbook = xlrd.open_workbook(file_path)
sheet = workbook.sheet_by_name(sheet_name)

# Create a new workbook to save the modified data
new_workbook = load_workbook(file_path)
new_sheet = new_workbook[sheet_name]

# Iterate through each cell in the sheet
for row_idx in range(sheet.nrows):
    for col_idx in range(sheet.ncols):
        cell = sheet.cell(row_idx, col_idx)
        # Check if the cell has red text color
        if cell.font.color and cell.font.color.rgb == 'FFFF0000':
            # Clear the cell value if it has red text color
            new_sheet.cell(row=row_idx+1, column=col_idx+1).value = None

# Save the modified workbook to a new file
new_workbook.save('E:\\_Projects\\FBOM\\modified_excel_file.xlsx')

# Load the modified Excel file into a DataFrame
df = pd.read_excel('E:\\_Projects\\FBOM\\modified_excel_file.xlsx', sheet_name=sheet_name, engine='openpyxl')

# Display the DataFrame
print(df)
```
