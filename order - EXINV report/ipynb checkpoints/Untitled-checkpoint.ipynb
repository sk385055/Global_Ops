```python
# Load the text file into a DataFrame
OSS = pd.read_csv(r'E:\_Projects\MCID Tagging\Input\OSS.txt', delimiter='|')  # Adjust the delimiter as needed


####################################################
# A-01-Order MCID update

OSS['CLASS']  = OSS['CLASS'].astype(str)

OSS = OSS[~OSS['INVENTORY_ITEM'].str.contains('%M%|-R000|F000|-C', regex=True)]

OSS = OSS.drop_duplicates(subset=['CLASS'], keep='first')

IA_01_OSS_02_Class = OSS[['CLASS']]
```

```python
import pandas as pd

# Assuming the data is loaded into DataFrames
EUM_OCP_IA = pd.read_excel(r'E:\_Projects\MCID Tagging\Input\OCP_IA_EUMOCP.xlsx')  # Replace with actual data loading

Country_mapping = pd.read_csv(r'E:\_Projects\MCID Tagging\Input\Country_mapping.csv') # Replace with actual data loading
TT_SIT_II = pd.read_csv(r'E:\_Projects\MCID Tagging\Output\LT.csv') # Replace with actual data loading
Order_MCID = pd.read_excel(r'E:\_Projects\MCID Tagging\Input\Order MCID.xlsx') # Replace with actual data loading
 # Replace with actual data loading





OSS = pd.read_csv(r'E:\_Projects\MCID Tagging\Input\OSS.txt', delimiter='|')  # Adjust the delimiter as needed

  # Replacercwith actual data l.xlsxoading

# First Query
IA_01_EUM_OCP_IA_Q = EUM_OCP_IA[EUM_OCP_IA['PID'].str.contains('M')]
IA_01_EUM_OCP_IA_Q = IA_01_EUM_OCP_IA_Q.merge(IA_01_OSS_02_Class, left_on='Class',right_on='CLASS', how='inner')


# Second Query
IA_06_OSS_country_mapping_update = IA_01_EUM_OCP_IA_Q.copy()
IA_06_OSS_country_mapping_update = IA_06_OSS_country_mapping_update.merge(Country_mapping, left_on='Whse', right_on='ORg', how='left')
#print(IA_06_OSS_country_mapping_update.columns)
IA_06_OSS_country_mapping_update['Whse_1'] = IA_06_OSS_country_mapping_update.apply(
    lambda row: row['Whse'] if pd.isnull(row['Country_code']) else row['Country_code'], axis=1)
IA_06_OSS_country_mapping_update['Expr1'] = IA_06_OSS_country_mapping_update.apply(
    lambda row: row['Region'] if pd.isnull(row['Region']) else row['Region'], axis=1)

# Third Query
TT_SIT_II['Class'] = TT_SIT_II['Class'].astype(str)
IA_06_OSS_country_mapping_update['Class'] = IA_06_OSS_country_mapping_update['Class'].astype(str)

IA_07_Country_sourcing = IA_06_OSS_country_mapping_update.merge(TT_SIT_II, left_on=['Whse_1', 'Class'], right_on=['Country Code', 'Class'], how='left')

IA_07_Country_sourcing.rename(columns={'Region_x': 'Region'}, inplace=True)
print(IA_07_Country_sourcing.columns)


IA_07_Country_sourcing = IA_07_Country_sourcing[['Plant', 'Class', 'Region', 'Whse']]

# Fourth Query
source_rule_local_update = IA_07_Country_sourcing.copy()
source_rule_local_update['Final_plant'] = source_rule_local_update.apply(
    lambda row: row['Plant'] if pd.notnull(row['Plant']) else row['Plant'], axis=1)
source_rule_local_update = source_rule_local_update[['Class', 'Region', 'Final_plant', 'Whse']]

# Fifth Query
OSS_restriction_feature_w_MCID = Order_MCID.merge(MCID_by_feature_TABLE, on='PID', how='left')
OSS_restriction_feature_w_MCID = OSS_restriction_feature_w_MCID.merge(OSS, left_on='PID', right_on='INVENTORY_ITEM', how='left')
OSS_restriction_feature_w_MCID = OSS_restriction_feature_w_MCID[
    ~OSS_restriction_feature_w_MCID['PID'].str.contains('R000|MC')]
OSS_restriction_feature_w_MCID = OSS_restriction_feature_w_MCID[
    ['PID', 'DESCRIPTION', 'MCID', 'Ennoconn', 'Budapest', 'Chennai (Dummy)', 'Chennai (Primary)', 
     'Chennai (Secondary)', 'Columbus', 'Dundee (NPI)', 'Easypoint', 'Jabil', 'Manaus', 
     'Midland', 'Sao Paulo', 'Shiloh (Dummy)', 'Waterloo']]



# Sixth Query

IA_01_EUM_OCP_IA_Q.drop_duplicates(inplace=True)
source_rule_local_update.drop_duplicates(inplace=True)
OSS_restriction_orders = IA_01_EUM_OCP_IA_Q.merge(source_rule_local_update, left_on=['Class', 'Whse'], right_on=['Class', 'Whse'], how='left')
OSS_restriction_orders = OSS_restriction_orders.merge(OSS_restriction_feature_w_MCID, left_on='PID', right_on='MCID', how='left')

# Apply the condition for the final query
def condition(row):
    if row['Final_plant'] == "CHE":
        return pd.isnull(row['Chennai (Primary)'])
    elif row['Final_plant'] == "CHS":
        return pd.isnull(row['Chennai (Secondary)'])
    elif row['Final_plant'] == "EC":
        return pd.isnull(row['Ennoconn'])
    elif row['Final_plant'] == "JAB":
        return pd.isnull(row['Jabil'])
    else:
        return True

OSS_restriction_orders = OSS_restriction_orders[OSS_restriction_orders.apply(condition, axis=1)]

# The resulting DataFrame is:
# OSS_restriction_orders
OSS_restriction_orders.to_csv(r'E:\_Projects\MCID Tagging\oss.csv',index=False)
print(OSS_restriction_orders)
```
