```python
import pandas as pd
import numpy as np
from pandas.tseries.offsets import BDay
```

```python
filename = 'dolphin_sco_20230831'
sheetname = 'Committed Ship Dates '
```

```python
import datetime
import pytz
from datetime import datetime, timedelta
d = datetime.now(pytz.timezone('Asia/Calcutta'))
filedate_str = d.strftime('%Y%m%d')
```

```python
data_jab_sco = pd.read_excel('E:\_Projects\orders/rcv/' + filename +'.xlsx', \
                             sheet_name=sheetname, engine='openpyxl', keep_default_na='', \
                            usecols='A:V')
```

```python
tmpl = pd.read_csv('E:\_Projects\orders\map\convert_format_jab_sco.csv', encoding='utf-8', low_memory=False, keep_default_na='')
```

```python
data_jab_sco.columns = data_jab_sco.columns.str.replace('\n',' ')
```

```python
cols_dt = ['SO CREATE', 'BUILD Wk', 'Req del date', 'SHIP DATE', 'Promise date', 'Wk NBD']
for col in data_jab_sco[cols_dt]:
    data_jab_sco[col] = pd.to_datetime(data_jab_sco[col], format='%Y-%m-%d', errors='coerce').dt.date
```

```python
# Fields for Open PO (common for Ennoconn, Jabil)
data_jab_sco['File'] = 'JAB'
data_jab_sco['SO'] = data_jab_sco['SALES ORDER']
data_jab_sco['Booked Dt'] = data_jab_sco['SO CREATE']
data_jab_sco['PID'] = data_jab_sco['MC']
data_jab_sco['Qty'] = data_jab_sco['Qty SHIP']
# data_jab_sco['PO'] = data_jab_sco['PO']
# data_jab_sco['Ship Dt'] = data_jab_sco['SHIP DATE']
data_jab_sco['Promise date'] = np.where(data_jab_sco['Promise date'].notnull(),pd.to_datetime(data_jab_sco['Promise date']) - BDay(3),data_jab_sco['Promise date'])
data_jab_sco['PD'] = data_jab_sco['Promise date']
data_jab_sco['Request Dt'] = data_jab_sco['Req del date']
data_jab_sco['Tot Qty'] = ''
data_jab_sco['Org PD'] = ''
data_jab_sco['Critical'] = data_jab_sco['CRITICAL']
data_jab_sco['Revenue'] = data_jab_sco['REVENUE']
data_jab_sco['Priority'] = data_jab_sco['PRIORITY']
data_jab_sco['Comments-1'] = data_jab_sco['NCR COMMENTS']
data_jab_sco['Comments-2'] = data_jab_sco['Jabil comments']
data_jab_sco['Customer'] = data_jab_sco['CUSTOMER']
```

```python

data_jab_sco['Demand'] = 'PO-JAB'
data_jab_sco['Supply Detail'] = data_jab_sco['PO']
data_jab_sco['DEMAND VAL'] = data_jab_sco['Supply Detail']
data_jab_sco['Supply Qty'] = data_jab_sco['Qty SHIP']
# data_jab_sco['PD_Order'] = data_jab_sco['PD_PO']
data_jab_sco['PD_PO'] = data_jab_sco['Promise date']

conditions = [data_jab_sco['PID'].str.contains('M')==True, data_jab_sco['PID'].str.contains('K')==True]
values = ['M','K']
data_jab_sco['PID Type'] = np.select(conditions, values, default='')

# data_jab_sco['Supply Qty'] = data_jab_sco['Qty']
data_jab_sco['Whse'] = 'EPL'
data_jab_sco['IO SSD'] = data_jab_sco['SHIP DATE']
data_jab_sco['Plant Ship date'] = data_jab_sco['SHIP DATE']
data_jab_sco['IO BOOKED DATE'] = data_jab_sco['Booked Dt']
data_jab_sco['NEED BY DATE'] = data_jab_sco['Request Dt']


# ---- Hierarchy SHIP DATE, PD_PO, Request Dt
data_jab_sco['Plant SSD'] = np.where(data_jab_sco['SHIP DATE'].notnull(), data_jab_sco['SHIP DATE'], \
                                     data_jab_sco['PD_PO'])
data_jab_sco['Plant SSD'] = np.where(data_jab_sco['Plant SSD'].isnull(), data_jab_sco['Request Dt'], \
                                     data_jab_sco['Plant SSD'])

data_jab_sco['Ship Dt'] = data_jab_sco['Plant SSD']

data_jab_sco['Validation Flag'] = np.where(data_jab_sco['PD_PO'].notnull(), 'PD OK', 'NPD')
data_jab_sco['Supply Type'] = np.where(data_jab_sco['PD_PO'].notnull(), 'PD', 'PO')
# data_jab_sco['PD_PO'] = np.where(data_jab_sco['SHIP DATE'].notnull(), data_jab_sco['SHIP DATE'], '')
data_jab_sco['Supplier'] = 'JABIL CIRCUIT INC - CHIHUAHUA'

condtn = [data_jab_sco['CLASS'].str[:4].isin(['7360']) & data_jab_sco['CLASS'].str.contains('N')==True, \
          data_jab_sco['CLASS'].str[:4].isin(['7360']) & data_jab_sco['CLASS'].str.contains('J')==True, \
          data_jab_sco['CLASS'].str[:4].isin(['7360']) & data_jab_sco['CLASS'].str.contains('0')==True]
vals = ['Narrow', 'Jarvis', 'Standard']
data_jab_sco['Prod Cate'] = np.select(condtn, vals, '')

data_jab_sco['Class'] = np.where(data_jab_sco['PID'].notnull(), data_jab_sco['PID'].str[:4], '')
data_jab_sco['Class'] = data_jab_sco['Class'].astype(str)
# data_jab_sco['Class'] = data_jab_sco['Class'].str[:4]

data_jab_sco = data_jab_sco[data_jab_sco['PID']!='']
data_jab_sco['Customer_End Customer'] = data_jab_sco['Customer']
# data_jab_sco.drop(columns=['Customer'], inplace=True)
```
