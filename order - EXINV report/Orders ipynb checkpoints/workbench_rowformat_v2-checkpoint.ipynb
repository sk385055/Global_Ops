```python
data2 = data[~data['IO No.'].isnull()]
```

```python
cols1 = ['DEMAND VAL', 'Org Code', 'Ship Date', 'SAD', 'Line Number']
data2 = data2.drop(cols1, axis=1)

cols2 = {'IO No.':'DEMAND VAL', 'Whse':'Org Code', 'IO SSD':'Ship Date', 'IO SAD':'SAD', 'IR No.':'Line Number'}
data2.rename(columns=cols2, inplace=True)
```

```python
data2['Demand'] = 'IO'
data2['Customer_End Customer'] = 'NCR INTERCOMPANY'
```

```python
frames = [data, data2]
result = pd.concat(frames, axis=0, ignore_index=True)
```

```python
region = pd.read_csv(r'E:\_Projects\lookups\Ref-IA-Region.csv', \
                    low_memory=False)
country = pd.read_csv(r'E:\_Projects\lookups\Ref-IA-Region-Map.csv', \
                    low_memory=False)
```

```python
result = pd.merge(result, region[['INVORG_', 'APAC','EMEA','AMM']], left_on=['SO Ship Org'], right_on='INVORG_', how='left')
result = pd.merge(result, country[['Country Code', 'Country Name','Region_']], left_on=['Org Code'], right_on='Country Code', how='left')
```

```python
import numpy as np
result['ds_apac'] = np.where(( (result['APAC']=='Y') | (result['APAC']=='PS') & (result['Region_']=='APAC')), 'APAC','')
result['ds_emea'] = np.where(( (result['EMEA']=='Y') | (result['EMEA']=='PS') & (result['Region_']=='EMEA')), 'EMEA','')
result['ds_amm'] = np.where(( (result['AMM']=='Y') | (result['AMM']=='PS') & (result['Region_'].isin(['NAMER','CLA']))), 'AMM','')
result.drop(columns={'INVORG_','APAC','EMEA','AMM','Country Code','Country Name','Region_'}, inplace=True)
```

```python
data_apac = result[result['ds_apac']=='APAC']
data_apac.drop(columns={'ds_apac', 'ds_emea', 'ds_amm'}, inplace=True)
data_apac.to_csv(r'E:\_Projects\_outputs\orders\snd\\' + file_name +file_date + '_io_row_apac_v2.csv', index=False)
# data_apac.to_csv(r'E:\_Projects\workbench\region\\' + file_name +file_date + '_io_row_apac_v2.csv', index=False)

data_emea= result[result['ds_emea']=='EMEA']
data_emea.drop(columns={'ds_apac', 'ds_emea', 'ds_amm'}, inplace=True)
data_emea.to_csv(r'E:\_Projects\_outputs\orders\snd\\' + file_name +file_date + '_io_row_emea_v2.csv', index=False)
# data_emea.to_csv(r'E:\_Projects\workbench\region\\' + file_name +file_date + '_io_row_emea_v2.csv', index=False)

data_amm= result[result['ds_amm']=='AMM']
data_amm.drop(columns={'ds_apac', 'ds_emea', 'ds_amm'}, inplace=True)
data_amm.to_csv(r'E:\_Projects\_outputs\orders\snd\\' + file_name +file_date + '_io_row_amm_v2.csv', index=False)
# data_amm.to_csv(r'E:\_Projects\workbench\region\\' + file_name +file_date + '_io_row_amm_v2.csv', index=False)
```
