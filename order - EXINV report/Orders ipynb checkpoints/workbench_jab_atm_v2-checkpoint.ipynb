```python
import datetime
import pytz
from datetime import datetime, timedelta
d = datetime.now(pytz.timezone('Asia/Calcutta'))
filedate_str = d.strftime('%Y%m%d')
```

```python
data_jab_atm = pd.read_excel('E:/_Projects/orders/rcv/' + filename + '.xlsx', \
                             sheet_name=sheetname, engine='openpyxl', keep_default_na='', \
                            usecols='A:AD')
## Change column name Qty and Bal and remove Shipment Date blank data


data_jab_atm= data_jab_atm[data_jab_atm['Ship Date'].notnull()]


data_jab_atm.rename(columns={'Qty' :'Bal1'},inplace = True)

data_jab_atm.rename(columns={'Bal':'Qty'},inplace = True)
data_jab_atm.rename(columns={'Bal1':'Bal'},inplace = True)

data_jab_atm.to_excel(r'E:/_Projects/orders/rcv/' + filename + '.xlsx',
                             sheet_name=sheetname,index=False)
```

```python
tmpl = pd.read_csv('./map/convert_format_jab_atm.csv', encoding='utf-8', low_memory=False, keep_default_na='')
```

```python
cols_dt = ['PO Receipt date', 'Build Wk', 'Ship Date', 'PD','Req.dlv.dt','8 Wks LT']
for col in data_jab_atm[cols_dt]:
    data_jab_atm[col] = pd.to_datetime(data_jab_atm[col], format='%Y-%m-%d', errors='coerce').dt.date

data_jab_atm['MCID'] = data_jab_atm['MCID'].astype(str)
```

```python
# Fields for Open PO (common for Ennoconn, Jabil)
data_jab_atm['File'] = 'JAB'
# data_jab_atm['SO'] = data_jab_atm['SO']
data_jab_atm['Booked Dt'] = data_jab_atm['PO Receipt date']
data_jab_atm['PID'] = data_jab_atm['MCID'].str.strip()
# data_jab_atm['Qty'] = data_jab_atm['Qty']
# data_jab_atm['PO'] = data_jab_atm['PO']
# data_jab_atm['Ship Dt'] = data_jab_atm['Ship Date']
# data_jab_atm['PD'] = data_jab_atm['PD']
data_jab_atm['Request Dt'] = data_jab_atm['Req.dlv.dt']
#data_jab_atm['Tot Qty'] = data_jab_atm['Tot. Req']
data_jab_atm['Tot Qty'] = data_jab_atm['Qty']    #Kathleen Requested to use BAL column as Qty 02-02-2024
#data_jab_atm['Tot Qty'] = data_jab_atm['Bal'] 
data_jab_atm['Org PD'] = ''
# data_jab_atm['Critical'] = data_jab_atm['Critical']
data_jab_atm['Revenue'] = data_jab_atm['Rev']
data_jab_atm['Priority'] = data_jab_atm['RI']
data_jab_atm['Comments-1'] = data_jab_atm['JABIL COMMENTS']
data_jab_atm['Comments-2'] = data_jab_atm['Adj Comment']
data_jab_atm['Customer'] = data_jab_atm['Customer']
data_jab_atm['Prod Cate'] = ''
```

```python
# Fields for Workbench
data_jab_atm['Demand'] = 'PO-JAB'
data_jab_atm['DEMAND VAL'] = data_jab_atm['PO']
data_jab_atm['Supply Detail'] = data_jab_atm['PO']
data_jab_atm['Supply Qty'] = data_jab_atm['Qty']   #Kathleen Requested to use BAL column as Qty 02-02-2024
#data_jab_atm['Supply Qty'] = data_jab_atm['Bal']
conditions = [data_jab_atm['PID'].str.contains('M')==True, data_jab_atm['PID'].str.contains('K')==True]
values = ['M','K']
data_jab_atm['PID Type'] = np.select(conditions, values, default='')
data_jab_atm['Whse'] = 'EPL'
# data_jab_atm['PD_PO'] = data_jab_atm['PD']
data_jab_atm['PD_PO'] = np.where(data_jab_atm['PD'].notnull(),pd.to_datetime(data_jab_atm['PD']) - BDay(3),data_jab_atm['PD'])
data_jab_atm['IO SSD'] = data_jab_atm['Ship Date']
data_jab_atm['Plant Ship date'] = data_jab_atm['Ship Date']
data_jab_atm['IO BOOKED DATE'] = data_jab_atm['Booked Dt']
data_jab_atm['Plant SSD'] = data_jab_atm['Ship Date']
data_jab_atm['NEED BY DATE'] = data_jab_atm['Request Dt']
data_jab_atm['Plant SSD'] = np.where(data_jab_atm['Ship Date'].notnull(), data_jab_atm['Ship Date'], \
                                     data_jab_atm['PD_PO'])
data_jab_atm['Plant SSD'] = np.where(data_jab_atm['Plant SSD'].isnull() , data_jab_atm['Request Dt'], \
                                     data_jab_atm['Plant SSD'])
data_jab_atm['Plant SSD'] = data_jab_atm['Plant SSD'].astype('datetime64[ns]')
data_jab_atm['Ship Dt'] = data_jab_atm['Plant SSD']
data_jab_atm['Validation Flag'] = np.where(data_jab_atm['PD_PO'].notnull(), 'PD OK', 'NPD')
data_jab_atm['Supply Type'] = np.where(data_jab_atm['PD_PO'].notnull(), 'PD', 'PO')
data_jab_atm['Supplier'] = 'JABIL CIRCUIT INC - CHIHUAHUA'
data_jab_atm['Customer_End Customer'] = data_jab_atm['Customer']
data_jab_atm = data_jab_atm[data_jab_atm['PID']!='']
```

```python

cndtn_pid1 = (data_jab_atm['PID'].str.contains('M','K') == False)
data_jab_atm['PID'] = np.where(cndtn_pid1,  \
                                 data_jab_atm['PID'].str[:3] + '-' + data_jab_atm['PID'].str[4:], \
                                 data_jab_atm['PID'])
data_jab_atm['Class'] = np.where(data_jab_atm['PID'].notnull() | data_jab_atm['PID'].str.len() >0, \
                                 data_jab_atm['PID'].str[:4], '')
```
