```python
if ver == 'EUM':
    f = ['emea','apac']
else:
    f=['amm']
    
```

```python
for file in f:

    data = pd.read_csv(r'E:\_Projects\_outputs\orders\snd\workbench_data_' + \
                           version + Date + '_io_row_' + file + '_v2.csv', encoding='utf-8', low_memory=False)
    print('workbench_data_'+version + Date + '_io_row_' + file + '_v2.csv')
    tmpl = pd.read_csv(r'.\map\convert_format_ia_' + file + '_v2.csv', encoding='utf-8', low_memory=False, keep_default_na='')
    lista = tmpl['file1'].tolist()
    listb = tmpl['file2'].tolist()
    df_sort = tmpl[["file2","order"]]


    def checkEmpty(s):
        try:
            return len(s)
        except:
            return False

    out = []
    x = len(lista)
    y = len(listb)
    for i in range(x):
        if not checkEmpty(lista[i]):
            lista[i] = 'x'+str(i)
            data['x'+str(i)] = ""

    for j in range(y):
        if checkEmpty(listb[j]):
            out.append(listb[j])
            data.rename(columns={lista[j]: listb[j]}, inplace=True)

    #sort
    df_sort = df_sort.dropna()
    df_sort["order"] = df_sort["order"].astype(int)
    df_sort = df_sort.sort_values("order")

    listc = df_sort["file2"].tolist()
    data = data[listc]
    
    for fld in ['Ship Date','Booked Dt','Request Dt','PD (Order)','SAD','CRSD','PRD']:
#         print(fld)
        data[fld] = pd.to_datetime(data[fld], format='%Y-%m-%d').dt.date
        
    data['Ship Date'] = np.where(data['Ship Date'].isnull(), '2030-12-31', data['Ship Date'])
    n = 79
    path = r'E:\_Projects\_outputs\orders\snd\IA-'+ver+'-'+file.upper()+'_'+ d

    data.iloc[:,:n].to_excel(path + '_v2.xlsx', index=False, na_rep='')
```
