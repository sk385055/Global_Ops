###### ####################### Find Sat & Sun ######### Vasanthan 

```python

# Ensure the date columns are datetime
OCP_IA['First Report date'] = pd.to_datetime(OCP_IA['First Report date'])
OCP_IA['FIO updated on'] = pd.to_datetime(OCP_IA['FIO updated on'])

# Function to check presence of Sat and Sun
def weekend_presence(start_date, end_date):
    if pd.isnull(start_date) or pd.isnull(end_date):
        return 0
    days = pd.date_range(start=start_date, end=end_date)
    has_saturday = any(day.weekday() == 5 for day in days)
    has_sunday = any(day.weekday() == 6 for day in days)
    return int(has_saturday) + int(has_sunday)

# Apply the function row-wise
OCP_IA['Weekend Count'] = OCP_IA.apply(
lambda row: weekend_presence(row['First Report date'], row['FIO updated on']),
axis=1
)

  
#############################################################################

OCP_IA['IO Ageing'] = (pd.to_datetime(OCP_IA['FIO updated on']) - pd.to_datetime(OCP_IA['First Report date'])).dt.days


OCP_IA['IO Ageing']  = OCP_IA['IO Ageing'] - OCP_IA['Weekend Count'] 

#######################################################################################################
# Create conditions and choices for ageing buckets
con1 = OCP_IA['IO Ageing'] < 0 #Less 10
con2 = OCP_IA['IO Ageing'] <= 5

OCP_IA['IO Buket'] = np.where(con1,'Early', \
 np.where(con2,'00-05 days','05+ days'))

OCP_IA['IO Buket'] = np.where((OCP_IA['FIO updated on'].isna()), 'Yet to be Schedule', OCP_IA['IO Buket'])

```
