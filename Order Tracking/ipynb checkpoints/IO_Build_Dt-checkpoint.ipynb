```python
Consolidation_Path = Path+'input\\Build_Dt\\Consolidation\\'


dfs = []

for file_name in os.listdir(Consolidation_Path):
    if file_name.endswith('.csv'):
        
        print(file_name)
        
        df = pd.read_csv(os.path.join(Consolidation_Path, file_name))
        dfs.append(df)

Consolidation = pd.concat(dfs, ignore_index=True)

#Consolidation.to_csv(Consolidation_Path+'Consolidate.csv',index=False)
```

```python
Data.rename(columns={'Actual Ship Date':'SHIPPED Dt'}, inplace=True)

#Data.to_csv(r'C:\Vasanthan\Report\Delivery Performance\Automation\Data.csv',index=False)
```

```python
# Convert each column to datetime format
date_columns = [
    'SHIPPED Dt',
    'IO Booked Dt(First)',
    'IO Booked Dt(Last)',
    'Booked Dt',
    'IO SSD (first)',
    'IO SSD (last)',
    'Manufactured Date'
]

for col in date_columns:
    #Data[col] = pd.to_datetime(Data[col], format='mixed')
    Data[col] = pd.to_datetime(Data[col])
```

```python
#Data.to_csv(r'C:\Vasanthan\Report\Delivery Performance\Automation\Data.csv',index=False)
```

### Build Date LT Caculation 

```python
Data['Build LT (first)'] = (Data['Manufactured Date'] - Data['IO SSD (first)']).dt.days

Data.loc[(Data['Build LT (first)'] < 1) | (Data['Build LT (first)'].isnull()), 'Compliance (first)'] = 'On-Time'

Data['Compliance (first)'] = Data['Compliance (first)'].fillna('Late')
```

```python
Data['Build LT (last)'] = (Data['Manufactured Date'] - Data['IO SSD (last)']).dt.days

Data.loc[(Data['Build LT (last)'] < 1) | (Data['Build LT (last)'].isnull()), 'Compliance (last)'] = 'On-Time'

Data['Compliance (last)'] = Data['Compliance (last)'].fillna('Late')
```

```python
def categorize_delivery(days):
    if days <= 0:
        return "On-Time"
    elif days <= 5:
        return "1) 1-5 Days Late"
    else:
        return "2) 30+ Days Late"

Data['Build LT (first) Bucket'] = Data['Build LT (first)'].apply(categorize_delivery)
Data['Build LT (last) Bucket'] = Data['Build LT (last)'].apply(categorize_delivery)
```

### Actual Ship Dt Calculation

```python
Data['Shp LT (first)'] = (Data['SHIPPED Dt'] - Data['IO SSD (first)']).dt.days

Data.loc[(Data['Shp LT (first)'] < 1) | (Data['Shp LT (first)'].isnull()), 'Shp Compliance (first)'] = 'On-Time'

Data['Shp Compliance (first)'] = Data['Shp Compliance (first)'].fillna('Late')
```

```python
Data['Shp LT (last)'] = (Data['SHIPPED Dt'] - Data['IO SSD (last)']).dt.days

Data.loc[(Data['Shp LT (last)'] < 1) | (Data['Shp LT (last)'].isnull()), 'Shp Compliance (last)'] = 'On-Time'

Data['Shp Compliance (last)'] = Data['Shp Compliance (last)'].fillna('Late')
```

```python
Data['Shp LT (first) Bucket'] = Data['Shp LT (first)'].apply(categorize_delivery)
Data['Shp LT (last) Bucket'] = Data['Shp LT (last)'].apply(categorize_delivery)
```
